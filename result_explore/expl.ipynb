```python
# 'winner'가 'buyer'인 경우를 승리로 간주하여 새로운 컬럼 추가
df['buyer_win'] = df['winner'] == 'buyer'

# 'buyer' 기준으로 'behaviour_2_persona' 별 승률 계산
win_rate = df.groupby('behaviour_2_persona')['buyer_win'].mean().reset_index()

# 승률을 백분율로 변환
win_rate['win_rate'] = win_rate['buyer_win'] * 100

# 결과 출력
print(win_rate[['behaviour_2_persona', 'win_rate']])
```

```python
def get_win_rate(df):
    # 'buyer' 기준으로 'behaviour_2_persona' 별 승률 계산
    win_rate = df.groupby('behaviour_2_persona')['buyer_win'].mean().reset_index()

    # 승률을 백분율로 변환
    win_rate['win_rate'] = win_rate['buyer_win'] * 100

    # 결과 출력
    print(win_rate[['behaviour_2_persona', 'win_rate']])
```

```python
for persona in df["behaviour_1_persona"].unique():
    temp = df.query(f"behaviour_2_persona == '{persona}'")
    get_win_rate(temp)
```

```python
pivot_table = pd.pivot_table(df, values='buyer_win', index='behaviour_2_persona', columns='behaviour_1_persona', aggfunc='mean', fill_value=0)
```
