```python
# Step 1: Load the Excel file
file_path = '/home/work/mgjeon/NegotiationArena/result_explore/face/persona_experiment_added_persona_item.xlsx'  # Adjust the file path if necessary
data = pd.read_excel(file_path, sheet_name='win_rate_by_model_persona')

# Step 2: Clean and reshape the data
melted_data = data.melt(id_vars=['Unnamed: 0'], var_name='Buyer', value_name='Performance')
melted_data.rename(columns={'Unnamed: 0': 'Seller'}, inplace=True)

# Step 3: Define the custom sorting orders for models and personas
model_order = ['gpt-3.5-turbo', 'gpt-4-mini', 'gpt-4', 'gpt-4-0']
persona_order = ['Altruistic', 'Cooperative', 'Control', 'Desperate', 'Selfish', 'Cunning', 'Competitive']

# Step 4: Function to extract the model and persona from the combined columns
def extract_model_persona(row):
    seller_parts = row['Seller'].split('_')
    buyer_parts = row['Buyer'].split('_')
    
    seller_model = '_'.join(seller_parts[1:])
    seller_persona = seller_parts[0]
    
    buyer_model = '_'.join(buyer_parts[1:])
    buyer_persona = buyer_parts[0]
    
    return pd.Series([seller_model, seller_persona, buyer_model, buyer_persona])

# Step 5: Clean data by dropping rows with NaN in 'Seller' or 'Buyer' columns
cleaned_data = melted_data.dropna(subset=['Seller', 'Buyer'])

# Step 6: Apply function to extract model and persona columns
cleaned_data[['Seller_model', 'Seller_persona', 'Buyer_model', 'Buyer_persona']] = cleaned_data.apply(extract_model_persona, axis=1)

# Step 7: Apply sorting based on the custom orders
cleaned_data['Seller_model'] = pd.Categorical(cleaned_data['Seller_model'], categories=model_order, ordered=True)
cleaned_data['Seller_persona'] = pd.Categorical(cleaned_data['Seller_persona'], categories=persona_order, ordered=True)
cleaned_data['Buyer_model'] = pd.Categorical(cleaned_data['Buyer_model'], categories=model_order, ordered=True)
cleaned_data['Buyer_persona'] = pd.Categorical(cleaned_data['Buyer_persona'], categories=persona_order, ordered=True)

# Step 8: Create the sorted pivot table
sorted_pivot_table = cleaned_data.pivot_table(index=['Seller_model', 'Seller_persona'], 
                                              columns=['Buyer_model', 'Buyer_persona'], 
                                              values='Performance', aggfunc='mean')

# Step 9: Display the sorted pivot table
print(sorted_pivot_table)
```

```python
import pandas as pd

# Step 1: Load the Excel file (Sheet1)
file_path = '/home/work/mgjeon/NegotiationArena/result_explore/face/persona_experiment_added_persona_item_v2.xlsx'  # Adjust the file path if necessary
data = pd.read_excel(file_path, sheet_name='Sheet1')

# Step 2: Filter relevant columns
data_filtered = data[['model_1', 'model_2', 'behaviour_1_persona', 'behaviour_2_persona', 'winner_model']].dropna()

# Step 3: Dynamically extract all unique model names from the data
unique_models = pd.concat([data_filtered['model_1'], data_filtered['model_2']]).unique()
print("Unique Models Found:", unique_models)

# Step 4: Define the custom sorting order dynamically based on the unique models in the data
model_order = sorted(unique_models)  # Sort models alphabetically or define a custom order if necessary
persona_order = ['Altruistic', 'Cooperative', 'Control', 'Desperate', 'Selfish', 'Cunning', 'Competitive']

# Step 5: Ensure there are no leading/trailing spaces in the model columns
data_filtered['model_1'] = data_filtered['model_1'].str.strip()
data_filtered['model_2'] = data_filtered['model_2'].str.strip()

# Step 6: Create a win/loss indicator
data_filtered['seller_win'] = data_filtered['winner_model'] == data_filtered['model_1']
data_filtered['buyer_win'] = data_filtered['winner_model'] == data_filtered['model_2']

# Step 7: Group by model and persona pairs and calculate win rates
grouped_data = data_filtered.groupby(
    ['model_1', 'behaviour_1_persona', 'model_2', 'behaviour_2_persona']
).agg(seller_wins=('seller_win', 'mean'),
       buyer_wins=('buyer_win', 'mean')).reset_index()

# Step 8: Debug step - Display the grouped data to ensure models are correct
print("Grouped Data (for Debugging):")
print(grouped_data.head())

# Step 9: Apply sorting based on the custom orders
grouped_data['model_1'] = pd.Categorical(grouped_data['model_1'], categories=model_order, ordered=True)
grouped_data['behaviour_1_persona'] = pd.Categorical(grouped_data['behaviour_1_persona'], categories=persona_order, ordered=True)
grouped_data['model_2'] = pd.Categorical(grouped_data['model_2'], categories=model_order, ordered=True)
grouped_data['behaviour_2_persona'] = pd.Categorical(grouped_data['behaviour_2_persona'], categories=persona_order, ordered=True)

# Step 10: Create the pivot table for seller and buyer win rates
pivot_table_seller = grouped_data.pivot_table(index=['model_1', 'behaviour_1_persona'],
                                              columns=['model_2', 'behaviour_2_persona'],
                                              values='seller_wins', aggfunc='mean')

pivot_table_buyer = grouped_data.pivot_table(index=['model_1', 'behaviour_1_persona'],
                                             columns=['model_2', 'behaviour_2_persona'],
                                             values='buyer_wins', aggfunc='mean')

# Step 11: Display the pivot tables
print("Seller Win Rates:")
print(pivot_table_seller)
print("\nBuyer Win Rates:")
print(pivot_table_buyer)
```
